```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time  


output_csv = 'books.csv'

# Initialize CSV file with headers
columns = ['title', 'price', 'rating']
pd.DataFrame(columns=columns).to_csv(output_csv, index=False)


for page in range(1,10):
    print(f"Scraping page {page}...")

    url = f'http://books.toscrape.com/catalogue/page-{page}.html'
    print(url)
    response = requests.get(url)

    if response.status_code != 200:
        print(f"Failed to retrieve page {page}")
        continue

    soup = BeautifulSoup(response.text, 'html.parser')
    containers = soup.find_all('article', class_='product_pod')

    data_full = []

    for container in containers:
        rating_tag = container.find('p', class_='star-rating')
        rating = rating_tag['class'][1]
        #print(rating)
        title_tag = container.find('a')
        title = title_tag.find('img')['alt']
        #print(title)
        price = container.find('p',class_ = 'price_color').text.strip()
        #print(price)
        data = {'title':title,'price':price,'rating':rating}
        #print(data)
        data_full.append(data)


    # Convert to DataFrame and append to CSV
    page_df = pd.DataFrame(data_full)
    page_df.to_csv(output_csv, mode='a', header=False, index=False)

    # Optional: sleep a bit between requests
    time.sleep(1)

print(f"Scraping complete. Data saved to {output_csv}")
```
